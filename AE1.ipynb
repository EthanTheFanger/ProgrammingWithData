## Question 5 - Book Rating System (Working With Files) 10 points
Design a new program that will inquire users about their favourite book and its rating.

exhibit a table of aggregated results. Sample of the table:

Favourite Book? Rating

Pride and Prejudice 5

1984 3


```
What is your favourite book? Pride and Prejudice
What is your rating for this book (1-5)? 5
Book and rating have been recorded!
Would you like to rate more books? no
Here are the accumulated ratings so far...
Pride and Prejudice - Rating 5
1984 - Rating 3
```

````python
import csv

#open Books.csv for writing
with open('Books.csv', 'w', newline='') as file:

    #create a csv writer object
    writer = csv.writer(file)

    #create the Favourite Book? and Rating heading
    writer.writerow(['Favourite Book', 'Rating'])

    #continously asking the user to rate the book
    while True:
        #ask the user for the favourite book 
        fav = input('WHat is your favourite book? ')

        #ask the user for the rating
        rating = input('What is your rating? ')

        #exception handling for invalid inputs
        #it would be possible with "try" and "except" but i found this method was the most succinct 
        while not rating.isnumeric() or not (1 <= int(rating) <= 5):
            rating = input('please enter a number between 1 and 5: ')
        
        #write the data to the csv file
        writer.writerow([fav, rating])
        print("Book and rating have been recorded!")

        #ask the user if they wish to continue
        is_Continue = input('Would you like to add another book? (y/n)').lower()

        #exception handling for invalid inputs
        while is_Continue != 'y' and is_Continue!= 'n':
            is_Continue = input('Please enter y or n: ').lower()

        #if the user says no, then break the loop and stop asking them
        if is_Continue == 'n':
            break

#print the data of the csv file after the loop
with open('Books.csv', 'r') as file:
    reader = csv.reader(file)

    #skip the header 
    next(reader)

    #printing out the songs 
    print("Here are the accumluated ratings so far: ")    
    for row in reader:
        print(f"{row[0]} - Rating {row[1]}")
````

day with no meal"


````python
#define reservation class
class Reservation:
    def __init__(self, guestName, eventLocation, eventDuration, mealOption):
        self.guestName = guestName
        self.eventLocation = eventLocation
        
        #if the input is not "half" or "full" then raise an error
        if eventDuration not in ("half", "full"):
            raise ValueError("eventDuration must be 'half' or 'full'")
            
        self.eventDuration = eventDuration
        self.mealOption = mealOption
    
    #announce reservation function
    def announceReservation(self):
        #how many days there are in event 
        duration_str = self.eventDuration
        
        #according to the example output, we might assume that when the eventduration is "full" then it is 3 days
        if self.eventDuration == "full":
            duration_str = "3 days"
        elif self.eventDuration == "half":
            duration_str = "1.5 days"

        #putting together the different attributes
        message = f"Greetings, I am {self.guestName} and I want to attend a conference in {self.eventLocation} for {duration_str} with {self.mealOption}"
        print(message)

#a reservation for a seminar, which extends the reservation class
class ConferenceReservation(Reservation):
    def __init__(self, guestName, eventLocation, eventDuration, mealOption="lunch included"):
        #if the input is not "half" or "full" then raise an error
        if eventDuration not in ("half", "full"):
            raise ValueError("eventDuration must be 'half' or 'full'")
            
        #putting the attributes into the superclass constructor 
        super().__init__(guestName, eventLocation, eventDuration, mealOption)
    
    
    #announce reservation function
    def announceReservation(self):
        #how many days there are in event 
        duration_str = self.eventDuration
        #according to the example output, we might assume that when the eventduration is "full" then it is 3 days
        if self.eventDuration == "full":
            duration_str = "3 days"
        elif self.eventDuration == "half":
            duration_str = "1.5 days"

        #putting together the different attributes
        message = f"Greetings, I am {self.guestName} and I want to attend a conference in {self.eventLocation} for {duration_str} with {self.mealOption}"
        print(message)

    
#a reservation for a seminar, which extends the reservation class
class SeminarReservation(Reservation):
    def __init__(self, guestName, eventLocation, eventDuration, mealOption="no meal"):
        #if the input is not "half" or "full" then raise an error
        if eventDuration not in ("half", "full"):
            raise ValueError("eventDuration must be 'half' or 'full'")
            
        #putting the attributes into the superclass constructor 
        super().__init__(guestName, eventLocation, eventDuration, mealOption)

    #announce reservation function
    def announceReservation(self):
        #how many days there are in event 
        duration_str = self.eventDuration
        #according to the example output, we might assume that when the eventduration is "full" then it is 3 days
        if self.eventDuration == "full":
            duration_str = "1 day"
        elif self.eventDuration == "half":
            duration_str = "half day"

        #putting together the different attributes
        message = f"Greetings, I am {self.guestName} and I want to attend a seminar in {self.eventLocation} for {duration_str} with {self.mealOption}"
        print(message)

# Instantiate attendees with different types of reservations
jennifer = ConferenceReservation("Jennifer", "New York", "full")
jennifer.announceReservation()

jack = SeminarReservation("Jack", "San Francisco", "half")
jack.announceReservation()
````
